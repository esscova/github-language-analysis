### obtendo dados do github

```python
import requests
import json
```

```python
# URL da API do GitHub para informações do usuário 'octocat'
url = "https://api.github.com/users/octocat"
```

```python
# fazendo uma requisicao
response = requests.get(url)

# verificando o status da resposta
response
```

```python
# convertendo a resposta em um dicionario
data = response.json()
data
```

```python
# imprimindo alguns dados do usuario
print(f"Nome de usuario: {data['login']}")
print(f"ID: {data['id']}")
print(f"Nome: {data['name']}")
print(f"URL: {data['html_url']}")
print(f"Repositórios públicos: {data['public_repos']}")
```

```python
# obter os repositórios de uma organização
data['repos_url']
```

```python
lista_repos = requests.get(data['repos_url']).json()
print(f"\nQuantidade de repositórios: {len(lista_repos)}\n")
print(json.dumps(lista_repos, indent=4))
```

```python
# nome do primeiro repositório
lista_repos[0]['name']
```

```python
# imprimindo o nome dos repositórios
for repo in range(len(lista_repos)): print(lista_repos[repo]['name'])
```

```python
# verificando se o repositório possui linguagem
lista_repos[0].get('languages', 'N/A')
```

```python
# verificando todos
for repo in range(len(lista_repos)):
    print(f"Repositório: {lista_repos[repo]['name']} - Linguagem: {lista_repos[repo].get('languages', 'N/A')}")
```

### dados da amazon

linguagens dos repos

```python
org_name = "amazon"
url_repos = f"https://api.github.com/orgs/{org_name}/repos"
params = {
    'per_page': 5,
    'page':1
}
try:
    response = requests.get(url_repos, params=params)
    response.raise_for_status()
    repos_data = response.json()
    for repo in repos_data: print(f" - {repo['name']} - ({repo.get('language', 'N/A')})")
except requests.exceptions.RequestException as e:
    print(f"Erro ao buscar  repositorios da organização {org_name} : {e}")
```
